```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import os

import missingno as msno
```

```python
df = pd.read_csv('./clean_ds.csv')
```

```python
plt.figure(figsize=(18,8))

colours = ['#34495E', 'seagreen'] 
sns.heatmap(df.isnull(), cmap=sns.color_palette(colours))
```

```python
col_list = df.columns.to_list()
col_list
```

```python
skills_df = df[['python', 'mysql',
 'postgresql',
 'tensorflow',
 'pytorch',
 'aws',
 'tableau',
 'powerbi',
 'hadoop',
 'kafka',
 'java',
 'excel',
 'scikit-learn',
 'sklearn',
 ' r ',
 'scala',
 'stata',
 'spss',
 'sas']]
```

```python
skills_df
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
df_transposed = skills_df.T
linked = linkage(df_transposed, method='ward')


plt.figure(figsize=(14, 10))  
dendrogram(linked,
           labels=df_transposed.index,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title('Hierarchical Clustering of Skills Dendrogram')
plt.xlabel('Skills')
plt.ylabel('Distance')
plt.xticks(rotation=45)  
plt.show()
```
